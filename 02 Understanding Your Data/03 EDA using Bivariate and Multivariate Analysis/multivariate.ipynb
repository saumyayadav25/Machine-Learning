##3. Box Plot(Numerical - Categorical)

```python
sns.boxplot(x='Sex',y='Age',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

##4. Distplot(Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'],hist=False)

# Plot the age distribution for passengers who survived
sns.distplot(titanic[titanic['Survived'] == 1]['Age'])
```

##5. HeadMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# titanic.groupby('Pclass').mean()['Survived']
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```
