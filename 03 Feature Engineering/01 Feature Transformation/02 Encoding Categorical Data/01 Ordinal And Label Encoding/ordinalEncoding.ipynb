```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
x_train
```

## Ordinal Encoding
->It is used for input columns

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# give categories in order of importance->for example: good to be assigned highest value and poor to ve assigned lowest value
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```

```python
x_train
```

```python
oe.categories_
```

## Label Encoding
-> It is used for output column

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```
